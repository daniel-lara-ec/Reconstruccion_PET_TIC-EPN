Este cuaderno incluye el ejemplo de aplicación de la retroproyección filtrada

# Módulos

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import radon, iradon
from skimage.data import shepp_logan_phantom
```

# Sinograma de un círculo

```python

imagen = plt.imread('Circulo.png')

imagen = imagen[:,:,0]

# invertimos la imagen
imagen = 1 - imagen

# Calcular el sinograma usando la transformación de Radon
theta = np.linspace(0., 180., max(imagen.shape), endpoint=False)
sinogram = radon(imagen, theta=theta)

# Visualizar la imagen original y el sinograma
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.imshow(imagen, cmap=plt.cm.Greys_r)
ax2.imshow(sinogram, cmap=plt.cm.Greys_r, extent=(0, 180, 0, sinogram.shape[0]), aspect='auto')
ax1.axis('off')
ax2.axis('off')

plt.tight_layout()
plt.show()
```

## Reconstrucción con 10 ángulos

```python
theta_recon = np.linspace(0., 10, max(imagen.shape), endpoint=False)
reconstruction_fbp = iradon(sinogram, theta=theta_recon, filter_name='ramp')

fig, ax3 = plt.subplots(1, 1, figsize=(12, 6))
ax3.imshow(reconstruction_fbp, cmap=plt.cm.Greys_r)
plt.axis('off')
plt.show()
```
