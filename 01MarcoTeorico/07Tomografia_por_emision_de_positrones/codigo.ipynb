# Módulos

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import radon
from skimage.data import shepp_logan_phantom
```

## Sinograma de un círculo

```python
imagen = plt.imread("Circulo.png")

imagen = imagen[:, :, 0]

imagen = 1 - imagen

theta = np.linspace(0.0, 180.0, max(imagen.shape), endpoint=False)
sinogram = radon(imagen, theta=theta)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.imshow(imagen, cmap=plt.cm.Greys_r)

ax2.imshow(
    sinogram, cmap=plt.cm.Greys_r, extent=(0, 180, 0, sinogram.shape[0]), aspect="auto"
)
ax1.axis("off")
ax2.axis("off")

plt.tight_layout()
plt.show()
```

## Sinograma de la forma de Shepp Logan

```python
image = shepp_logan_phantom()

theta = np.linspace(0., 180., max(image.shape), endpoint=False)
sinogram = radon(image, theta=theta)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.imshow(image, cmap=plt.cm.Greys_r)
ax2.imshow(sinogram, cmap=plt.cm.Greys_r, extent=(0, 180, 0, sinogram.shape[0]), aspect='auto')

ax1.axis('off')
ax2.axis('off')

plt.tight_layout()
plt.show()
```
