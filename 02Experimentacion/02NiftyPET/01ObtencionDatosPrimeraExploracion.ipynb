```python
import pickle
import pandas as pd
```

```python
from matplotlib.pyplot import (
    matshow,
    colorbar,
    xlabel,
    ylabel,
    plot,
    legend,
    grid,
)

import matplotlib.pyplot as plt
```

```python
def recortar(x):
    return x[100:-120, 110:-110]

def recorta2(x):
    return x[15:,95:-110]

def recorta3(x):
    return x[25:,120:-120]
```

```python
# Cargar las variables desde el archivo pickle
with open('datos_resultados/resultados_metrica_completa/recon_mlem_inicial.pickle', 'rb') as file:
    reconstruccion = pickle.load(file)
```

```python
ruta = 'datos_resultados/exploracion_inicial/'
```

```python

for i in range(len(reconstruccion)):

    matshow(recortar(reconstruccion[i][75,:,:]), cmap='magma')
    plt.axis('off')
    plt.savefig(f'{ruta}transversal_mlem_{i*10}.png')

    matshow(recorta3(reconstruccion[i][:,175,:]), cmap='magma')
    plt.axis('off')
    plt.savefig(f'{ruta}sagitalmlem_{i*10}.png')

    matshow(recorta2(reconstruccion[i][:,:,175]), cmap='magma')
    plt.axis('off')
    plt.savefig(f'{ruta}coronal_mlem_{i*10}.png')
```

```python
# Cargar las variables desde el archivo pickle
with open('datos_resultados/resultados_metrica_completa/datos_iteraciones_inicial.pickle', 'rb') as file:
    datos_reconstruccion = pickle.load(file)

datos_reconstruccion = datos_reconstruccion[1:] 
```

```python
datos_reconstruccion
```

```python
datos_ssim = [ {'ssim' : x['SSIM'], 'iteracion' : x['iteracion']} for x in datos_reconstruccion]
```

```python
df = pd.DataFrame(datos_ssim)
```

```python
df.to_csv('datos_resultados/exploracion_inicial/ssim_mlem.csv', index=False)
```

```python
# plot datos_ssim logatimico
plt.plot(df['ssim'], label='SSIM', color=(0,0,1))
plt.yscale('logit')


# save in svg
plt.savefig('SSIM_mlem_inicial.svg')
plt.show()


```
