Este cuaderno incluye los ejemplos de las geometrías de adquisición presentados en la sección 3.2.1

# Módulos

```python
from cil.framework import AcquisitionGeometry
from cil.utilities.display import show_geometry
```

## Geometría 2D paralela

```python
ag = AcquisitionGeometry.create_Parallel2D(detector_position=[0,0])
ag.set_panel(num_pixels=10)
ag.set_angles(angles=range(0,180))

show_geometry(ag)

print(ag)
```

## Geometría paralela 3D

```python
# Definimos una geometría 3D con detector en el origen, de 10x15 píxeles y 180 ángulos.
ag_3d = AcquisitionGeometry.create_Parallel3D(detector_position=[0,0,0])
ag_3d.set_panel(num_pixels=[10,15])
ag_3d.set_angles(angles=range(0,180))

# Mostramos la figura que ilustra la geometría
show_geometry(ag_3d)

# Mostramos la información detallada de la geometría
print(ag_3d)
```

## Geometría de abanico 2D

```python
ag_fan_beam = AcquisitionGeometry.create_Cone2D(source_position=[0,-5],detector_position=[0,5])
ag_fan_beam.set_panel(num_pixels=5)
ag_fan_beam.set_angles(angles=range(0,180))
    
show_geometry(ag_fan_beam)
```

## Geometría de cono 3D

```python
ag_cone_beam = AcquisitionGeometry.create_Cone3D(source_position=[0,-10,0],detector_position=[0,10,0])
ag_cone_beam.set_panel(num_pixels=[10,10])
ag_cone_beam.set_angles(angles=range(0,180))
    
show_geometry(ag_cone_beam)
```
